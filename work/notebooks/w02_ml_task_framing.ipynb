## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
# ML Task Type: Classification
#
# My lane (Ranking Signal Analysis) explores which content and search signals
# relate to a page's performance. I'm framing this as a classification problem:
# predicting whether a page is "declining" or "not declining" (a binary label),
# using observable signals (search volume, position, CTR, word count, age, etc.)
# as inputs.
#
# Classification fits because the outcome is naturally binary (declining vs.
# not), and it lets me directly test which signals actually help separate
# declining pages from stable/growing ones - which signals the model finds
# useful to make this prediction ARE the signals that relate to performance,
# which is exactly what my Lane 1 research question is asking.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
# Target / Proxy: is_declining_label
#
# Target: is_declining_label = 1 if trend_direction == "down", else 0.
#
# This label comes from a defined rule based on an observed outcome
# (trend_direction), not a hand-picked guess - trend_direction itself reflects
# real recent performance data. It's a proxy in the sense that it captures
# the CURRENT trend bucket rather than a future outcome, so it's a starting
# point rather than the strongest possible label (a stronger version would
# use a future window, e.g. declining over the NEXT 30 days).
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
# Success Metric: Precision@50
#
# Precision@50 = of the top 50 pages the model flags as "declining," what
# fraction are actually declining?
#
# I'm choosing this over overall accuracy because in practice, a content team
# can only review a limited number of pages (say, the top 50) - so what matters
# is whether the model's TOP picks are correct, not how well it scores every
# page in the dataset. This also matches the metric already used in the
# starter notebooks, where a good model clearly beat a simple hand-written
# rule on this exact measure (baseline ~0.24 vs random forest ~0.74).
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# Why ML beats a fixed rule here:
#
# A simple hand-written rule can capture obvious cases (e.g. "flag pages that
# are stale AND still getting traffic"), but real page performance depends on
# many signals interacting together - search volume, position, CTR, word count,
# age, competition level - and these relationships aren't simple thresholds.
#
# This was already demonstrated in the starter pipeline: a hand-written baseline
# rule scored Precision@50 = 0.240, while a trained random forest model scored
# 0.740 - roughly 3x better. The model could combine many weaker signals
# together in ways a human-written if/else rule can't easily replicate, because
# the "right" combination of thresholds isn't obvious or fixed - it has to be
# learned from the data itself.
#
# In short: individual signals are noisy and interact in non-obvious ways, so
# a rule based on one or two conditions misses patterns that a model, which
# can weigh and combine many signals at once, is able to find.
```
