## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os, subprocess

REPO_URL = "https://github.com/saleharehan97-del/Flyrank-ML-internship"
REPO_DIR = "Flyrank-ML-internship"

if not os.path.isdir(REPO_DIR):
    subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)

os.chdir(REPO_DIR)
print("Working dir:", os.getcwd())

import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Number 1: correlation between search volume and actual traffic
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")

# Number 2: CTR gap between top position and deep pages
ctr_by_tier = df.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print("\nAverage CTR by position tier:")
print(ctr_by_tier.round(3))

# Number 3: how many pages are declining vs not
declining_rate = (df["trend_direction"] == "down").mean()
print(f"\nShare of pages currently declining: {declining_rate:.3f}")
```
